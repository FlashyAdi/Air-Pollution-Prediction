```python


import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

print("Upload your city_day.csv file")
uploaded = files.upload()

filename = list(uploaded.keys())[0]
df = pd.read_csv(filename)

df.columns = df.columns.str.strip().str.lower()

print("\nFirst 5 rows:")
print(df.head())

print("\nColumns and types:")
print(df.info())

print("\nMissing values per column:")
print(df.isnull().sum())

print("\nBasic statistics:")
print(df.describe())

plt.figure(figsize=(8,5))
plt.hist(df['pm2.5'].dropna(), bins=30, color='skyblue')
plt.title("PM2.5 Distribution")
plt.xlabel("PM2.5")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(8,5))
plt.hist(df['aqi'].dropna(), bins=20, color='lightgreen')
plt.title("AQI Distribution")
plt.xlabel("AQI")
plt.ylabel("Frequency")
plt.show()

print("\n[MODEL BUILDING PLACEHOLDER]")
print("Here you can add regression/classification model to predict AQI or pollutants.")
```
