```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files

uploaded = files.upload()
filename = next(iter(uploaded))
df = pd.read_csv(filename)

print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
display(df.head())

df.columns = df.columns.str.strip().str.lower()
df['date'] = pd.to_datetime(df['date'], errors='coerce')

print("\nMissing values BEFORE:")
print(df.isna().sum().sort_values(ascending=False))

num_cols = df.select_dtypes(include=['number']).columns
df[num_cols] = df[num_cols].apply(lambda c: c.fillna(c.mean()))

print("\nMissing values AFTER (numeric cols filled):")
print(df.isna().sum().sort_values(ascending=False))

print("\nBasic statistics:")
display(df.describe())

if 'aqi' in df.columns:
    plt.figure(figsize=(8,5))
    plt.hist(df['aqi'].dropna(), bins=30, edgecolor="black")
    plt.title("AQI Distribution")
    plt.xlabel("AQI"); plt.ylabel("Frequency")
    plt.show()

if 'aqi' in df.columns and 'city' in df.columns:
    sample_city = df['city'].mode()[0]
    city_data = df[df['city'] == sample_city].dropna(subset=['date']).sort_values('date')
    plt.figure(figsize=(12,5))
    plt.plot(city_data['date'], city_data['aqi'])
    plt.title(f"AQI Trend Over Time – {sample_city}")
    plt.xlabel("Date"); plt.ylabel("AQI")
    plt.tight_layout()
    plt.show()

plt.figure(figsize=(12,8))
corr = df[num_cols].corr()
sns.heatmap(corr, annot=False, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap (Numeric Features)")
plt.show()

if 'aqi' in df.columns and 'city' in df.columns:
    plt.figure(figsize=(12,5))
    df.groupby('city')['aqi'].mean().sort_values(ascending=False).head(10).plot(kind='bar')
    plt.title("Top 10 Most Polluted Cities (Avg AQI)")
    plt.ylabel("Average AQI")
    plt.tight_layout()
    plt.show()

df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month

if 'city' in df.columns:
    df['city_code'] = df['city'].astype('category').cat.codes

print("\nPreview after preprocessing:")
display(df.head())

df.to_csv("city_day_processed_week2.csv", index=False)
print("\nSaved: city_day_processed_week2.csv")
```
